```python
dm=df_dupr[df_dupr.sex == "M"].dDUPR
dF=df_dupr[df_dupr.sex == "F"].dDUPR
sm=df_dupr[df_dupr.sex == "M"].sDUPR
sf=df_dupr[df_dupr.sex == "F"].sDUPR
dms=df_dupr.loc[df_dupr.sex=="M"][lambda x:pd.notna(x.sDUPR)]["dDUPR"]
dfs=df_dupr.loc[df_dupr.sex=="F"][lambda x:pd.notna(x.sDUPR)]["dDUPR"]
da=df_dupr.dDUPR
ds=df_dupr.sDUPR
dds = df_dupr.loc[lambda x:pd.notna(x.sDUPR)]["dDUPR"]
```

```python
df_raw = pd.DataFrame()
dict_metrx = {}
df_quant = pd.DataFrame()
mlist = [da,ds,dds,dm,dF,sm,sf,dms,dfs]
nlist = ["Doubles","Singles","Doubles w/ Singles","Men Doubles","Women Doubles","Men Singles","Women Singles","Men Doubles w/ Singles","Women Doubles w/ Singles"]
for n,m in zip(nlist,mlist):
    df_raw[n] = m
    dict_metrx[n] = {"mean":m.mean().round(3),"median":m.median().round(3),"count":sum(pd.notna(m))}
    df_quant[n] = m.quantile([.01*x for x in range(0,101)],interpolation="nearest")
df_metrx=pd.DataFrame(dict_metrx.values())
df_metrx["metric"]=dict_metrx.keys()
df_quant.reset_index(drop=False,inplace=True,names="pctile")
```

```python
df_metrx = df_metrx[["metric","mean","median","count"]]
df_metrx.to_csv("../docs/local_dupr_stats.csv")
df_metrx
```

```python
df_quant.to_csv("../docs/local_dupr_percentiles.csv")
```
